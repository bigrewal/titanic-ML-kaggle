```python
train_X = data_x[:713, :]
train_Y = data_y[:713]

dev_X = data_x[713:, :]
dev_Y = data_y[713:]

m_train,_ = train_X.shape
m_dev,_ = dev_X.shape
```

```python
# Let's start building our DL model using Tensorflow
def create_placeholders(n_f):
    X = tf.placeholder(tf.float32, [None, n_f])
    Y = tf.placeholder(tf.float32,[None,1])
    
    return X,Y
```

```python
def initialize_parameters():
    W1 = tf.get_variable("W1", [8,5], initializer = tf.contrib.layers.xavier_initializer())
    b1 = tf.get_variable("b1", [5], initializer = tf.zeros_initializer())
    
    W2 = tf.get_variable("W2", [5,1], initializer = tf.contrib.layers.xavier_initializer())
    b2 = tf.get_variable("b2", [1], initializer = tf.zeros_initializer())
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2 }
    
    return parameters
```

```python
def forward_propagation(X, parameters):

    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    
    Z1 = tf.add(tf.matmul(X,W1), b1)                                              
    A1 = tf.sigmoid(Z1)                                              
    Z2 = tf.add(tf.matmul(A1,W2), b2)                                           
    A2 = tf.sigmoid(Z2)                                              
                                                 
    ### END CODE HERE ###
    
    return A2
```

```python
def compute_cost(A2,Y):
    cost = -tf.reduce_mean(Y * tf.log(A2) + (1 - Y) * tf.log(1 - A2))
    
    return cost
```

```python
def model(train_X, train_Y, tot_ittr = 5000, learning_rate = 0.001, print_cost = True):
    
    m, n_f = train_X.shape
    
    # Create placeholders
    X, Y = create_placeholders(n_f)
    
    # Initialise weights and biases
    parameters = initialize_parameters()
    
    # Create forward propogation graph in tensorflow
    A2 = forward_propagation(X, parameters)
    
    # Add cost function to tensorflow graph
    cost = compute_cost(A2,Y)
    
    optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(cost)
    
    # Initialize all the variables
    init = tf.global_variables_initializer()

    # Start the session to compute the tensorflow graph
    with tf.Session() as sess:
        
        # Run the initialization
        sess.run(init)
        
        for ittr in range(tot_ittr):
            loss,_ = sess.run([cost,optimizer], feed_dict={X: train_X, Y: train_Y})
            
            if print_cost == True and ittr % 100 == 0:
                print ("Cost after ittr %i: %f" % (ittr, loss))
                
    
    # lets save the parameters in a variable
        parameters = sess.run(parameters)
        print ("Parameters have been trained!")
    
    # Calculate the correct predictions
        correct_prediction = tf.equal(tf.round(A2), Y)

        # Calculate accuracy on the test set
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))

        print ("Train Accuracy:", accuracy.eval({X: train_X, Y: train_Y}))
        print ("Dev Set Accuracy:", accuracy.eval({X: dev_X, Y: dev_Y}))
    
    return parameters 
```

```python
parameters = model(train_X, train_Y)
```

```python
#Run the model on the test set
test_data = test.drop("PassengerId", axis=1).copy()
test_data = test_data.as_matrix().astype(np.float32)

sess = tf.Session()
# Run the variables initialization (if needed), run the operations
#result = sess.run([forward_propagation], feed_dict = {X: test_data})

result = sess.run(forward_propagation(test_data, parameters))
print(result[0:5,])
result = np.round(result)
result = np.squeeze(np.asarray(result)).astype(np.int32)
print(result.dtype)
sess.close()

submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": result
    })
submission.to_csv('titanic-ML-kaggle/submission.csv', index=False)
```
